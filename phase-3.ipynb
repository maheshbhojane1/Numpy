```python
import numpy as np
import matplotlib as plt
```

## Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

```python
sales_data = np.array([
    [1, 150000, 180000, 120000, 250000], 
    [2, 160000, 200000, 240000, 260000],
    [3, 180000, 190000, 230000, 270000],
    [4, 170000, 210000, 250000, 290000],
    [5, 190000, 220000, 250000, 280000],
])

print("Top three restaurant data: \n", sales_data[:, 1:])
```

```python
# total sales per year
year_total = np.sum(sales_data[:, 1:], axis=0)
sums = np.sum(year_total[:])
print(year_total)
print("All company revenu sum: ",sums)


min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

max_sales = np.max(sales_data[:, 1:], axis=1)
print(max_sales)

avrage_sales = np.mean(sales_data[:, 1:], axis=1)
print(avrage_sales)

cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(f"\n{cumsum}")
```

## Vectors

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Addition", vector1 + vector2)
print("Multiplication", vector1 * vector2)
print("Dot product", np.dot(vector1, vector2))
```
