## Numpy Array and Basics

```python
import numpy as np 
```

## Creating array from list

```python
arr_1d = np.array([1, 2, 3, 4])
print("arr_1d: ", arr_1d)

arr_2d = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print("arr_2d: ", arr_2d)
```

## List vs numpy

```python
py_list = [1, 2, 3]
print("Python list multi", py_list * 2)

numpy_array = np.array([1, 2, 3])
print("numpy array multi", numpy_array * 2)


import time

start = time.time()
py_list = [i * 2 for i in range(1000000)]
print("\n list operation time: ", time.time() - start)


start = time.time()
np_array = np.arange(1000000) * 2 
print("\n Numpy operation time: ", time.time() - start)
```

## Creating array from scrach

```python
zeros = np.zeros((3, 4))
print("Zeros array: \n", zeros)

ones = np.ones((2, 3))
print("Onse array: \n", ones)

full = np.full((2, 3), 4)
print("full array: \n", full)

random = np.random.random((2, 3))
print("random array: \n", np.round(random))

sequence = np.arange(0, 10, 2)
print("sequence array: \n", sequence)
```

## Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("Vector: ", vector)

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix: ", matrix)

tensor = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("tensor: ", tensor)
```

## Array Properties

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Shape: ", arr.shape)
print("Dimension: ", arr.ndim)
print("Size: ", arr.size)
print("DataType", arr.dtype)
```

## Array Reshaping

```python
arr = np.arange(12)
print("Original array: ", arr)

reshaped = arr.reshape((3, 4))
print("Reshaped: ", reshaped)

flattened = reshaped.flatten()
print("Flatten: ",flattened)

# revel (return view, instead of copy)
reveled = reshaped.ravel()
print("Reveled: ", reveled)

transpose = reshaped.T
print("Transpose: ", transpose)
```
